# Analysis of **Airbnb** April 2020 data for New York City

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('D:/College stuff/DataScientistNanodegree/blog/listings.csv')
df.head()
```

# which Neighbourhood Group has most of the listings?

```python
status_vals =df.neighbourhood_group.value_counts() 

((status_vals*50000)/df.shape[0]).plot(kind="bar",color='orangered').legend();
plt.title("Where are most of the staying options in New York?");
```

```python
status_vals =df.neighbourhood.value_counts() 

((status_vals[:10])/df.shape[0]).plot(kind="bar",color='orangered').legend();
plt.title("Where are most of the staying options in New York?");
```

# Where can I find the most expensive and most cheap stay?

```python
neighbour_room_df = df[['neighbourhood_group', 'price', 'room_type']].groupby(['neighbourhood_group']).mean()
neighbour_room_df.sort_values('price',inplace=True)

neighbour_room_df.plot.bar(figsize=(20,7),color='orangered').legend(fontsize=20);
plt.title("Where can I find the most expensive and most cheap stay?");
```

# Which type of accomodation to prefer?

```python
df[['room_type', 'price']].groupby(['room_type']).mean().plot.bar(figsize=(20,5),color='orangered',fontsize=10).legend(fontsize=20);
plt.title("Which type of accommodation to  prefer?");
```
